# Import diabetes.csv dataset 

```python
df = pd.read_csv("diabetes.csv")
df
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
df = df.apply(lambda x: x if x.name == 'Pregnancies' else x.replace(0, np.nan))
```

```python
df
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
medi_pre, medi_gluc, medi_bp, medi_skin, medi_ins, medi_bmi = df['Pregnancies'].median(), df['Glucose'].median(), df['BloodPressure'].median(), df['SkinThickness'].median(), df['Insulin'].median(), df['BMI'].median()
```

```python
df['Pregnancies'].fillna(medi_pre, inplace=True)
df['Glucose'].fillna(medi_gluc, inplace=True)
df['BloodPressure'].fillna(medi_bp, inplace=True)
df['SkinThickness'].fillna(medi_skin, inplace=True)
df['Insulin'].fillna(medi_ins, inplace=True)
df['BMI'].fillna(medi_bmi, inplace=True)
```

```python
df.isnull().sum()
```

# Visualize Distributions

```python
import seaborn as sb
sb.pairplot(df, hue='Outcome')
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({"Diabetic":1,"Non Diabetic":0})
```

```python
sb.heatmap(df.corr(), annot = True, cmap='coolwarm')
plt.show()
```

```python
df
```

# Divide the data into input and output

```python
X = df.drop('Outcome', axis=1)
```

```python
y = df['Outcome']
```

# Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)
```

# Create Model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42, max_depth=100)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model.fit(X_train, y_train)
```

# Display Decision Tree

```python

%matplotlib qt
tree.plot_tree(model)
```

# Predict the x_test 

```python
y_pred = model.predict(X_test)
```

# Display Training Accuracy

```python
model.score(X_train, y_train)
```

# Display Test Accuracy

```python
model.score(X_test, y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python

from sklearn.metrics import classification_report, roc_auc_score, accuracy_score
print(classification_report(y_test, y_pred))
```

# Save Model State

```python
import pickle
```
